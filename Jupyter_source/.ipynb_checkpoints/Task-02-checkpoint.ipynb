# 4. 按月份统计每个月论文的提交数量

```python
import pandas as pd
import networkx as nx

data = pd.read_pickle('../../arxiv-metadata-oai-2019.pkl')
data
```

```python
data['Month'] = data['update_date'].map(lambda x: x.split('-')[1])
data['Month'].astype(int)
data.groupby('Month').size().to_frame('Number of papers').plot()
```

```python
G = nx.Graph()
for row in data.iloc[:5].itertuples():
    submitter = row.submitter
    authors = row.authors_parsed
    authors = [' '.join(reversed(au)) for au in authors]
    print(authors)
    for author in authors:
        G.add_edge(submitter, author)
nx.draw(G, with_labels=True)
```
