```python
from json import loads
import pandas as pd


# 不是完整的 json,是一个 json 集合。 不能直接使用 pd.read_json ,所以创建函数读取数据

def readJsonFile(path, column=['id', 'submitter', 'authors', 'title', 'comments', 'journal-ref', 'doi',
                               'report-no', 'categories', 'license', 'abstract', 'versions',
                               'update_date', 'authors_parsed'], col_num=170000):
    data = []
    with open(path) as f:
        for line, json in enumerate(f):
            if line == col_num:
                break
            data.append({col: loads(json)[col] for col in column})
    return pd.DataFrame(data)


df = readJsonFile('../arxiv-metadata-oai-2019.json')
df
```

# 1. 论文摘要和论文标题出现关键词的频率

```python
taget1_df = readJsonFile('../arxiv-metadata-oai-2019.json', column=['id', 'title', 'abstract'])
taget1_df
```

```python
# 定义统计函数
def word_count(txt):
    word_dic = {}
    for ch in '''!"$%&()*+,-./;:<=>?@[\\]^_{|}~''\n\t ''':
        txt = txt.replace(ch, ' ')
    for word in txt.split(' '):
        word = word.lower()
        if word_dic.get(word):
            word_dic[word] += 1
        else:
            word_dic[word] = 1
    return word_dic


# 使用停词库过滤掉部分无用词
def wordsfilter(word_dic):
    stopwords = []
    # 读取停词库
    with open('./stopwords.txt') as f:
        for word in f:
            stopwords.append(word.split('\n')[0])
    for word in list(word_dic.keys()):  # 不能在遍历字典的同时修改字典, 所以转换成 list
        if word in stopwords:
            del word_dic[word]
    # 字典可以使用空字符串来作为 key 所以需要特殊处理一下
    del word_dic['']


# worldcloud 库 实现词云图
def creat_wordcloudImg(word_dic, file_name, word_num=50):
    import wordcloud
    img = wordcloud.WordCloud(background_color='white', font_path='msyh.ttc', width=2000, height=1500,
                              max_words=word_num).generate_from_frequencies(word_dic)
    img.to_file(f'./{file_name}.png')


def word_count2df(column):
    word_dic = {}
    for i in range(len(taget1_df)):
        for word in word_count(taget1_df[column][i]):
            if word_dic.get(word):
                word_dic[word] += 1
            else:
                word_dic[word] = 1
    wordsfilter(word_dic)
    # 排序
    # print(sorted(word_dic.items(), key=lambda x: x[1], reverse=True))
    # 创建词云图
    creat_wordcloudImg(word_dic, column, word_num=200)

    words_data = {'word': [], 'amount': []}
    for t in word_dic:
        words_data['word'].append(t)
        words_data['amount'].append(word_dic[t])
    return pd.DataFrame(words_data).sort_values(by='amount', ascending=False)


titleword_df = word_count2df('title')
titleword_df
```

```python
abstract_df = word_count2df('abstract')
abstract_df
```

# 3.论文在 arxiv 系统中所属的类别统计

```python
taget3_df = readJsonFile('../arxiv-metadata-oai-2019.json', column=['id', 'categories'])
taget3_df
# print(taget3_df['categories'][i])
```

```python
categorie_num = {}


# 根据第三个级别的分类简写找出它所在的大类
def find_level1(level3, leveltabel):
    if len(leveltabel[leveltabel['categories'] == level3].values) == 0:
        level1 = leveltabel[leveltabel['archive_id'] == level3]['group_name']
    else:
        level1 = leveltabel[leveltabel['categories'] == level3]['group_name']
    if len(level1) == 0:
        return 'Not Exit'
    else:
        return level1.values[0]


leve_tabel = pd.read_csv('./categorys.csv')
for i in range(len(taget3_df)):
    for ctgr in taget3_df['categories'][i].split(' '):
        level1 = find_level1(ctgr, leve_tabel)
        if categorie_num.get(level1):
            categorie_num[level1] += 1
        else:
            categorie_num[level1] = 1
categorie_num
```

```python
group_df = {'group': [], 'number': []}
for group in categorie_num:
    group_df['group'].append(group)
    group_df['number'].append(categorie_num[group])
pd.DataFrame(group_df)
```
